```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.usinflationcalculator.com/inflation/consumer-price-index-and-annual-percent-changes-from-1913-to-2008/"
url_req = requests.get(url).text

soup = BeautifulSoup(url_req, 'html')
print(type(soup))
```

```python
tables = soup.find_all( "table")
print("There are {} tables".format(len(tables)))
```

```python
print(tables)
```

```python
print(soup.title.text)
```

```python
data = tables[0].tbody.find_all('tr')
df_data = [[td.getText().strip() for td in tr.find_all('td')] for tr in data]
df_data
```

```python
df_columns = ["Year", "Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sep", "Oct", "Nov", "Dec", "Avg", "Dec-Dec", 
              "Avg-Avg"]
```

```python
df1 = pd.DataFrame(df_data, columns = df_columns)
df1
```

```python
df1.drop(df1.index[[0,1]])
```

```python
df1.to_csv("CPI_1913_2021.csv", index=False)
```
